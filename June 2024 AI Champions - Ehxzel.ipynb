```python
#Summarizing numerical and categorical variables
iris = pd.read_csv("Iris.csv")
print(iris.PetalLengthCm, iris.PetalWidthCm)
```

```python
#Visualizing the distributions between columns
# import pandas as pd
# from matplotlib import pyplot as plt

plt.plot(iris.Species, iris.PetalWidthCm, color = "DarkCyan", label = "PetalWidth")
plt.plot(iris.Species, iris.SepalWidthCm, color ="tomato", label = "SepalWidth")
plt.title("Visualization of Iris Dataset")
plt.xlabel("Species")
plt.ylabel("Petal Width in Cm")
plt.style.use("fivethirtyeight")

plt.legend()
plt.show()
print(iris.head())
```
